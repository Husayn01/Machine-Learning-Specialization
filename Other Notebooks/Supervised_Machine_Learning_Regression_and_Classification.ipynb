```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x_train = np.array([1.0, 2.0])
y_train = np.array([300.0, 500.0])
print(f"x_train = {x_train}")
print(f"y_train = {y_train}")
```

```python
print(f"x_train.shape = {x_train.shape}")
m = x_train.shape[0]
print(f"Number of training examples is: m = {m}")
```

```python
# m is the number of training examples
m = len(x_train)
print(f"Number of training examples is: {m}")
```

```python
i = 0
x_i = x_train[i]
y_i = y_train[i]
print(f"(x^({i}), y^({i})) = ({x_i}, {y_i})")
```

```python
plt.scatter(x_train, y_train, marker='x', c='r')
plt.title("Housing Prices")
plt.ylabel('Price (in 1000s of dollars)')
plt.xlabel('Size (1000 sqft)')
plt.show()
```

```python
w = 100
b = 100
print(f"w: {w}")
print(f"b: {b}")
```

```python
def compute_model_output(x, w, b):
    """
    Computes the prediction of a linear model
    Args:
      x (ndarray (m,)): Data, m examples
      w,b (scalar)    : model parameters
    Returns
      f_wb (ndarray (m,)): model prediction
    """
    m = x.shape[0]
    f_wb = np.zeros(m)
    for i in range(m):
        f_wb[i] = w * x[i] + b

    return f_wb
```

```python
tmp_f_wb = compute_model_output(x_train, w, b,)

# Plot our model prediction
plt.plot(x_train, tmp_f_wb, c='b',label='Our Prediction')

# Plot the data points
plt.scatter(x_train, y_train, marker='x', c='r',label='Actual Values')

# Set the title
plt.title("Housing Prices")
# Set the y-axis label
plt.ylabel('Price (in 1000s of dollars)')
# Set the x-axis label
plt.xlabel('Size (1000 sqft)')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x_train = [1.0, 2.0]
y_train = [300.0, 500.0]
```

```python
def predicted(x, w, b):
  y_pred = []
  m = len(x)
  for i in range(m):
    f_wb = w * x[i] + b
    y_pred.append(f_wb)
  return y_pred
```

```python
predicted(x_train, 200, 100)
```

```python
plt.plot(x_train, predicted(x_train, 200, 100), c='b')
plt.scatter(x_train, y_train, marker='x', c='r')
```

```python
def compute_cost(x, y, w, b):
  cost_sum = 0
  m = len(x)
  for i in range(m):
    f_wb = w * x[i] + b
    cost = (f_wb - y[i]) ** 2
    cost_sum += cost
  total_cost = (1 / (2 * m)) * cost_sum
  return total_cost
```

```python
w_vals = []
for i in range(0, 401, 5):
  w_vals.append(i)
cost_vals = []
for w in w_vals:
  cost = compute_cost(x_train, y_train, w, 100)
  cost_vals.append(cost)
```

```python
plt.figure(figsize=(8, 5))
plt.plot(w_vals, cost_vals, 'b')
plt.xlabel('w')
plt.ylabel('Cost J(w)')
plt.title('Cost vs w')
plt.grid(True)
plt.show()
```
